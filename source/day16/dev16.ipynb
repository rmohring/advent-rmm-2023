# Part 1

```python
%%time
# Start top left, go right
beams = [((0,-1),(0,1))]  
seen_beams = set()

energized = set()
energized.add( (0,0) )

loopctr = 0
grid = g.grid

while len(beams)>0:
    loopctr += 1
    if loopctr % 10 == 0:
        print(f"Working loop {loopctr}...beams = {len(beams)}...answer = {len(energized)}")
        
    #print(beams)
    beams_new = []
    for ctr, ((vrt,hrz),(v_vrt,v_hrz)) in enumerate(beams):
        pos = (vrt+v_vrt, hrz+v_hrz)
        if ( pos[0]>=grid.shape[0] or pos[1]>=grid.shape[1] or pos[0]<0 or pos[1]<0):
            continue

        energized.add(pos)
        
        cell = grid[pos[0], pos[1]]
        
        #print(f"LOOP ### {loopctr:04d} ### Beam {ctr} cell: {pos} v_vrt{v_vrt:2d} v_hrz{v_hrz:2d} # {cell} #")
        # Split horiz
        if cell=="|":
            if v_vrt==0:
                b = [(pos, (1,0)), (pos, (-1,0))]
            elif v_hrz==0:
                b = [(pos, (v_vrt,0))]
            
        # Split vert
        if cell=="-":
            if v_hrz==0:
                b = [(pos, (0,1)), (pos, (0,-1))]
            elif v_vrt==0:
                b = [(pos, (0,v_hrz))]
        # Mirrors
        if cell=="/":
            if v_vrt==0:
                b = [(pos, (-v_hrz,0))]
            elif v_hrz==0:
                b = [(pos, (0,-v_vrt))]
        if cell=="\\":
            if v_vrt==0:
                b = [(pos, (v_hrz,0))]
            elif v_hrz==0:
                b = [(pos, (0,v_vrt))]
        if cell==".":
            b = [(pos, (v_vrt,v_hrz))]
        
        b_unique = [x for x in b if x not in seen_beams]
        seen_beams.update(b_unique)
        beams_new.extend(b_unique)
        
    beams = beams_new.copy()
```

# Part 2

```python
def run_test(beams):
    seen_beams = set()
    energized = set()

    loopctr = 0
    grid = g.grid

    while len(beams)>0:
        loopctr += 1
#         if loopctr % 10 == 0:
#             print(f"Working loop {loopctr}...beams = {len(beams)}...answer = {len(energized)}")

        #print(beams)
        beams_new = []
        for ctr, ((vrt,hrz),(v_vrt,v_hrz)) in enumerate(beams):
            pos = (vrt+v_vrt, hrz+v_hrz)
            if ( pos[0]>=grid.shape[0] or pos[1]>=grid.shape[1] or pos[0]<0 or pos[1]<0):
                continue

            energized.add(pos)

            cell = grid[pos[0], pos[1]]

            #print(f"LOOP ### {loopctr:04d} ### Beam {ctr} cell: {pos} v_vrt{v_vrt:2d} v_hrz{v_hrz:2d} # {cell} #")
            # Split horiz
            if cell=="|":
                if v_vrt==0:
                    b = [(pos, (1,0)), (pos, (-1,0))]
                elif v_hrz==0:
                    b = [(pos, (v_vrt,0))]

            # Split vert
            if cell=="-":
                if v_hrz==0:
                    b = [(pos, (0,1)), (pos, (0,-1))]
                elif v_vrt==0:
                    b = [(pos, (0,v_hrz))]
            # Mirrors
            if cell=="/":
                if v_vrt==0:
                    b = [(pos, (-v_hrz,0))]
                elif v_hrz==0:
                    b = [(pos, (0,-v_vrt))]
            if cell=="\\":
                if v_vrt==0:
                    b = [(pos, (v_hrz,0))]
                elif v_hrz==0:
                    b = [(pos, (0,v_vrt))]
            if cell==".":
                b = [(pos, (v_vrt,v_hrz))]

            b_unique = [x for x in b if x not in seen_beams]
            seen_beams.update(b_unique)
            beams_new.extend(b_unique)

        beams = beams_new.copy()
    
    return len(energized)
```

```python
%%time
results = []

grid_height = g.grid.shape[0]
grid_width = g.grid.shape[1]

for i in range(grid_height):
    start_beam = [((i,-1),(0,1))]
    results.append( (run_test(start_beam), start_beam) )
    start_beam = [((i,grid_width),(0,-1))]
    results.append( (run_test(start_beam), start_beam) )

for i in range(grid_width):
    start_beam = [((-1,i),(1,0))]
    results.append( (run_test(start_beam), start_beam) )
    start_beam = [((grid_height,i),(-1,0))]
    results.append( (run_test(start_beam), start_beam) )

for x in reversed(sorted(results)[-15:]):
    print(x)
```
