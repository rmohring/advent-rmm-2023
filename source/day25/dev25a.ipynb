```python
%%time
#h = open("input25-1.dat")
fh = open("test.dat")
#fh = open("test2.dat")

lines = fh.readlines()
lines = [x.strip("\n") for x in lines]

class Graph:
    def __init__(self):
        self.edges = set()
        self.nodes = {}

    def process_str(self, initstr):
        spl = initstr.split(": ")
        source = spl[0]
        ends = spl[1].split(" ")
        for e in ends:
            self.add_edge( (source, e) )

    def add_edge(self, edge_tup):
        self.edges.add( edge_tup )

    def finalize(self):
        for (a,b) in self.edges:
            if a not in self.nodes:
                self.nodes[a] = set()
            self.nodes[a].add(b)
            if b not in self.nodes:
                self.nodes[b] = set()
            self.nodes[b].add(a)


results = {}
paths = {}

def traverse(this_node, start_node, seen=None, level=0, graph=None):
    level += 1
    local_seen = seen.copy()
    local_seen.append(this_node)
    
    # Go to node and pick a new option
    for next_node in graph.nodes[this_node]:
        if next_node in local_seen:
            continue

        #print(start_node, local_seen, next_node)

        if level < results.get((start_node, next_node),999999):
            results[(start_node, next_node)] = level
            paths[(start_node, next_node)] = local_seen.copy() + [next_node]
        
        traverse(next_node, start_node, seen=local_seen, level=level, graph=graph)       
    return
    
g = Graph()
for line in lines:
    g.process_str(line)

g.finalize()


for i,n in enumerate(g.nodes):
    print(f"Working start node: {n} {i+1}/{len(g.nodes)}...")
    seen = [] #[n]
    traverse(n, n, seen=seen, level=0, graph=g)
    raise

# The results will have the longest paths... I'm guessing this path will have to traverse at least one bridge
#q = pd.Series(results)
#big_start, big_end = q[q==q.max()].head(1).index.tolist()[0]

ctr = Counter()
for junk, nodes in paths.items():
    for pair in list(zip(nodes, nodes[1:])):
        ctr[tuple(sorted(list(pair)))] += 1
trim_edges = pd.Series(ctr).sort_values(ascending=False).head(3).index.tolist()
print("TRIMMING:", trim_edges)
new_g = Graph()
for edge in g.edges:
    if (edge not in trim_edges) and ( (edge[1], edge[0]) not in trim_edges):
        new_g.add_edge(edge)
new_g.finalize()

# Finally, traverse from one side of one of the trimmed edges and count nodes:
seen = [] 
results = {}
paths = {}
traverse(trim_edges[0][0], trim_edges[0][0], seen=seen, level=0, graph=new_g)

total = set()
for v in paths.values():
    total.update(v)
print(len(total), len(g.nodes))

# print("Calculating eccentricities...")
# eccentricity = {}
# for n in g.nodes:
#     eccentricity[n] = max([ecc for (a,b),ecc in results.items() if a==n])
#     print(n, eccentricity[n])

# print("Calculating edge ecc sums...")
# ecc_sum = {}
# for (n1,n2) in g.edges:
#     #print(n1,n2)
#     ecc_sum[(n1,n2)] = eccentricity[n1] + eccentricity[n2]
```
